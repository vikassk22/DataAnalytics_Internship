```python
import pandas as pd
df = pd.read_csv("../data/TMDB_movie_dataset.csv", encoding="ISO-8859-1")
print(f"Dataset has {df.shape[0]} rows and {df.shape[1]} columns.")
df.head(5)
```

```python
print(df.columns.tolist())
print(df.dtypes)
```

```python
print("\nMissing values:\n", df.isnull().sum())
df.describe(include="all").T.head(20)
```

```python
print("\nDuplicate values:\n", df.duplicated().sum())
df.describe(include="all").T.head(20)
```

```python
print(f"Filtered out movies released before 2000")
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce', dayfirst=True)
df['release_year'] = df['release_date'].dt.year
filtered_df = df[df['release_year'] >= 2000]
print(f"We have {filtered_df.shape[0]} movies released in or after 2000.")
```

```python
filtered_df[['budget', 'revenue', 'runtime', 'popularity']].describe().T
```

```python
print(filtered_df['genres'].value_counts().head(10))
print(filtered_df['original_language'].value_counts().head(10))
```

```python
filtered_df['release_year'].value_counts().sort_index().plot(kind='bar', figsize=(12,4))
```

```python
filtered_df.plot.scatter(x='budget', y='revenue', alpha=0.3)
```

```python
import matplotlib.pyplot as plt
import collections

genre_list = filtered_df['genres'].dropna().str.split('|').sum()
genre_counts = dict(collections.Counter(genre_list))

pd.Series(genre_counts).sort_values(ascending=False).head(10).plot(kind='bar', figsize=(10,4))
plt.ylabel("Number of Movies")
plt.title("Top 10 Genres (2000 onwards)")
plt.show()
```
